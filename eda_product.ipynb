```python
# category 
 # 3. PRODUCT CATEGORY ANALYSIS
print("\n3. PRODUCT CATEGORY ANALYSIS")
print("-" * 40)
    
# Extract product categories from names
def categorize_product(name):
    name_lower = name.lower()
    if any(word in name_lower for word in ['shampo', 'shampoo', 'dove','clear', 'pepsodent','lux','vaseline','rexona','sabun','deodorant']):
        return 'Personal Care'
    elif any(word in name_lower for word in ['pureit', 'pure it', 'water purifier', 'filter']):
        return 'Water Purifier'
    elif any(word in name_lower for word in ['bango', 'buavita', 'juice','blueband']):
        return 'FnB'
    else:
        return 'Others'
    
df['category'] = df['name'].apply(categorize_product)
print("Product Categories:")
category_counts = df['category'].value_counts()
print(category_counts)
    
print(f"\nAverage Price by Category:")
category_price = df.groupby('category')['price'].agg(['mean', 'count']).round(0)
print(category_price)
```

```python
import sqlite3
import pandas as pd
import os
from datetime import datetime
```

```python


def insert_csv_to_sqlite(csv_file='products_data.csv', db_file='products_item.db', table_name='products'):
    """
    Insert data dari CSV file ke SQLite database
    """
    
    print("🚀 Starting CSV to SQLite import process...")
    print("=" * 60)
    
    # STEP 1: Check if CSV file exists
    if not os.path.exists(csv_file):
        print(f"❌ Error: CSV file '{csv_file}' not found!")
        print("📝 Please make sure the file exists in the current directory.")
        return False
    
    try:
        # STEP 2: Read CSV file
        print(f"📖 Reading CSV file: {csv_file}")
        df = pd.read_csv(csv_file)
        print(f"✅ CSV loaded successfully!")
        print(f"   • Total rows: {len(df):,}")
        print(f"   • Total columns: {len(df.columns)}")
        print(f"   • Columns: {list(df.columns)}")
        
        # STEP 3: Data validation and cleaning
        print(f"\n🧹 Data cleaning and validation...")
        
        # Check for missing values
        missing_data = df.isnull().sum()
        if missing_data.sum() > 0:
            print(f"⚠️  Missing values found:")
            for col, missing in missing_data.items():
                if missing > 0:
                    print(f"   • {col}: {missing} missing values")
            
            # Fill missing values
            df = df.fillna({
                'name': 'Unknown Product',
                'price': 0,
                'originalprice': 0,
                'discountpercentage': 0.0,
                'rating': 0.0,
                'detail': '',
                'platform': 'unknown',
                'productmasterid': '',
                'createdat': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
            })
            print(f"✅ Missing values handled!")
        
        # Add createdat if not exists
        if 'createdat' not in df.columns:
            df['createdat'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
            print(f"✅ Added 'createdat' column with current timestamp")
        
        # Data type conversions
        print(f"🔧 Converting data types...")
        df['price'] = pd.to_numeric(df['price'], errors='coerce').fillna(0).astype(int)
        df['originalprice'] = pd.to_numeric(df['originalprice'], errors='coerce').fillna(0).astype(int)
        df['discountpercentage'] = pd.to_numeric(df['discountpercentage'], errors='coerce').fillna(0.0)
        df['rating'] = pd.to_numeric(df['rating'], errors='coerce').fillna(0.0)
        
        print(f"✅ Data types converted successfully!")
        
        # STEP 4: Connect to SQLite database
        print(f"\n🔗 Connecting to SQLite database: {db_file}")
        conn = sqlite3.connect(db_file)
        cursor = conn.cursor()
        print(f"✅ Database connection established!")
        
        # STEP 5: Create table with proper schema
        print(f"\n📋 Creating table: {table_name}")
        
        create_table_sql = f'''
        CREATE TABLE IF NOT EXISTS {table_name} (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT NOT NULL,
            price INTEGER NOT NULL,
            originalprice INTEGER NOT NULL,
            discountpercentage REAL NOT NULL,
            rating REAL NOT NULL,
            detail TEXT,
            platform TEXT NOT NULL,
            productmasterid TEXT,
            createdat TEXT
        )
        '''
        
        cursor.execute(create_table_sql)
        conn.commit()
        print(f"✅ Table '{table_name}' created/verified successfully!")
        
        # STEP 6: Check existing data
        cursor.execute(f"SELECT COUNT(*) FROM {table_name}")
        existing_count = cursor.fetchone()[0]
        print(f"📊 Existing records in table: {existing_count:,}")
        
        # STEP 7: Insert data using pandas to_sql (most efficient)
        print(f"\n💾 Inserting data into SQLite database...")
        
        # Remove 'id' column if it exists (let SQLite auto-increment)
        if 'id' in df.columns:
            df_to_insert = df.drop('id', axis=1)
        else:
            df_to_insert = df.copy()
        
        # Insert data
        df_to_insert.to_sql(table_name, conn, if_exists='append', index=False)
        
        # STEP 8: Verify insertion
        cursor.execute(f"SELECT COUNT(*) FROM {table_name}")
        new_count = cursor.fetchone()[0]
        inserted_count = new_count - existing_count
        
        print(f"✅ Data insertion completed!")
        print(f"   • Records inserted: {inserted_count:,}")
        print(f"   • Total records now: {new_count:,}")
        
        # STEP 9: Data verification and sample display
        print(f"\n🔍 Data verification...")
        
        # Basic statistics
        cursor.execute(f'''
        SELECT 
            COUNT(*) as total_products,
            AVG(price) as avg_price,
            MIN(price) as min_price,
            MAX(price) as max_price,
            COUNT(DISTINCT platform) as platforms_count
        FROM {table_name}
        ''')
        
        stats = cursor.fetchone()
        total, avg_price, min_price, max_price, platforms = stats
        
        print(f"📊 DATABASE STATISTICS:")
        print(f"   • Total Products: {total:,}")
        print(f"   • Average Price: Rp {avg_price:,.0f}")
        print(f"   • Price Range: Rp {min_price:,} - Rp {max_price:,}")
        print(f"   • Number of Platforms: {platforms}")
        
        # Platform distribution
        cursor.execute(f"SELECT platform, COUNT(*) FROM {table_name} GROUP BY platform ORDER BY COUNT(*) DESC")
        platform_stats = cursor.fetchall()
        
        print(f"\n🏪 PLATFORM DISTRIBUTION:")
        for platform, count in platform_stats:
            percentage = (count / total) * 100
            print(f"   • {platform.title()}: {count:,} products ({percentage:.1f}%)")
        
        # Sample data display
        print(f"\n📝 SAMPLE DATA (First 5 records):")
        cursor.execute(f"SELECT name, price, platform, rating FROM {table_name} LIMIT 5")
        sample_data = cursor.fetchall()
        
        for i, (name, price, platform, rating) in enumerate(sample_data, 1):
            rating_stars = "⭐" * int(rating) if rating > 0 else "No rating"
            print(f"   {i}. {name[:50]}...")
            print(f"      Price: Rp {price:,} | Platform: {platform} | Rating: {rating_stars}")
        
        # STEP 10: Create useful indexes for performance
        print(f"\n⚡ Creating indexes for better performance...")
        
        indexes = [
            f"CREATE INDEX IF NOT EXISTS idx_{table_name}_platform ON {table_name}(platform)",
            f"CREATE INDEX IF NOT EXISTS idx_{table_name}_price ON {table_name}(price)",
            f"CREATE INDEX IF NOT EXISTS idx_{table_name}_rating ON {table_name}(rating)",
        ]
        
        for index_sql in indexes:
            cursor.execute(index_sql)
        
        conn.commit()
        print(f"✅ Indexes created successfully!")
        
        # Close connection
        conn.close()
        print(f"\n🎉 IMPORT COMPLETED SUCCESSFULLY!")
        print(f"   Database: {db_file}")
        print(f"   Table: {table_name}")
        print(f"   Records: {inserted_count:,} inserted")
        
        return True
        
    except Exception as e:
        print(f"❌ Error during import: {str(e)}")
        if 'conn' in locals():
            conn.close()
        return False

# ADVANCED FUNCTIONS FOR DATABASE OPERATIONS
def query_products_by_platform(db_file='products_item.db', platform=None):
    """Query products by specific platform"""
    conn = sqlite3.connect(db_file)
    
    if platform:
        query = "SELECT * FROM products WHERE platform = ? ORDER BY price DESC"
        df = pd.read_sql_query(query, conn, params=[platform])
    else:
        query = "SELECT * FROM products ORDER BY price DESC"
        df = pd.read_sql_query(query, conn)
    
    conn.close()
    return df

```

```python
def get_platform_statistics(db_file='products_item.db'):
    """Get comprehensive statistics by platform"""
    conn = sqlite3.connect(db_file)
    cursor = conn.cursor()
    
    query = '''
    SELECT 
        platform,
        COUNT(*) as product_count,
        AVG(price) as avg_price,
        MIN(price) as min_price,
        MAX(price) as max_price,
        AVG(rating) as avg_rating,
        AVG(discountpercentage) as avg_discount,
        COUNT(CASE WHEN rating > 0 THEN 1 END) as rated_products
    FROM products 
    GROUP BY platform
    ORDER BY product_count DESC
    '''
    
    df = pd.read_sql_query(query, conn)
    conn.close()
    return df

def export_platform_data_to_csv(db_file='products_item.db', platform='tokopedia'):
    """Export specific platform data back to CSV"""
    df = query_products_by_platform(db_file, platform)
    filename = f'{platform}_products.csv'
    df.to_csv(filename, index=False)
    print(f" {platform.title()} data exported to {filename}")
    return filename
```
